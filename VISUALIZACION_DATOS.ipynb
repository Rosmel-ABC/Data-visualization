#VISUALIZACION DE DATOS

```python
# importando librerias para la visualización
import matplotlib.pyplot as plt
import seaborn as sns


import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)  # 100 numeros del 0 al 10
y = np.sin(x)    # funcion seno
z = np.cos(x)   # funcion coseno

fig, ax = plt.subplots(figsize = (10,3))  # fig = figura (lienzo), ax = eje, figsize = tamañp del grafico

# Esto es acumulativo  es decir estarán en el mismo lienzo ambos gráficos
ax.plot(x , y , linewidth = 5)   #variables para los ejes , linewidth = grosor de linea
ax.plot(x,z , color = "green" , linestyle="dashed") #color = color de linea  , dashed = linea punteada

plt.title("Mi primer gráfico")  #titulo del gráfico
plt.xlabel("Eje X") #nombre del eje
plt.ylabel("Eje Y")  #nombre del eje

plt.show() # Mostrar gráfico
```

## Gráfico de dispersión

```python
edad = np.random.random_integers(18 , 65 , 100)         # array de numpy de 100 valores entee 18 y 65
ingresos = np.random.random_integers(1050,40000, 100)   # array  de 100 valores ente 1 050 y 40 000

fig, ax = plt.subplots()  # lienxo y ejes

ax.scatter(edad , ingresos)    # gráfico de dispersión
plt.title("Relación edad vs ingresos")   # titulo del gráfico
plt.xlabel("Edad")       #nombre del eje x
plt.ylabel("Ingresos")   #nombre del eje y 
plt.show()               #mostar gráficos
```

## Gráfico de Barras

```python
pais = ["Perú" , "Colombia" , "Argentina" , "Chile"]    #lista de paises
pbi = [1000 ,  800 ,  700 , 600]           # lista de pbi  ( datos artificiales )

fig , ax= plt.subplots() # lienzo y ejes

ax.bar(pais, pbi)  #gráfico de barras

plt.show()   #mostar gráfico
```

## Histogramas

```python
x = np.random.normal(100, 6 , 10000)  # datos ( promedio , desviación estandar , cantidad de datos)
plt.hist(x , bins = 1000)  # histograma

plt.title("Distribución Normal")   # titulo del gráfico
plt.show()                         #mostrar gráfico
```

## Gráfico Forcasting

```python
years = np.arange (2000 ,  2024 , 1)
ganancias = np.random.random_integers(10000, 50000, len(years))

limite_superior = ganancias *1.25
limite_inferior = ganancias*0.75

fig , ax = plt.subplots()                 # lienzo y ejes
ax.fill_between(years , limite_inferior, limite_superior, alpha= .4)   # gráfico
ax.plot(years , ganancias)       # gráfico de la linea
plt.title("Forecating")         #titulo del gráfico
plt.show()   #mostrar gráfico
```

```python
x = np.linspace(0,20,1000)
y = np.sin(x)    # funcion seno
z = np.cos(x)    # funcion coseno
t = np.tan(x)    # funcion tangente
normal=np.random.normal(0,1,1000)      # array de 1000 numeros entre 0 y 1

fig,axs = plt.subplots(nrows = 2 , ncols = 2 , figsize = (15,7))  # fig = figura (lienzo), ax1 , ax2 = ejes, figsize = tamañp del grafico
# nrowa = filas  y ncols = columnas
#figsize = tamaño kienzo


#axs[fila, columna]
axs[0,0].plot(x,y , linewidth = 5)   #variables para los ejes , linewidth = grosor de linea
axs[0,1].plot(x,z , color = "green" , linestyle="dashed") #color = color de linea  , dashed = linea punteada
axs[1,0].plot(x,t)
axs[1,1].hist(normal)


axs[0,0].set_title("Función Seno")  #nombre de grafico 1
axs[0,1].set_title("Función Coseno") # nombre grafico 2

fig.suptitle("Mi primer gráfico" , fontsize = 20)  #Titulo General

plt.xlabel("Eje X") #nombre del eje
plt.ylabel("Eje Y")  #nombre del eje
plt.show() # primer gráfico
```

##Gráfico Polar

```python
r = np.arange(0,2, 0.01)

theta = 2*np.pi*r
fig, ax = plt.subplots( subplot_kw ={"projection": "polar"} , figsize = ( 10,5))

ax.plot(theta, r)
plt.show()
```
